```python
numbers = []
for i in range(5):
    num = int(input(f"Введите число {i+1}: "))
    numbers.append(num)

min_number = min(numbers)
max_number = max(numbers)

print(f"Наименьшее число = {min_number}")
print(f"Наибольшее число = {max_number}")
```

```python
def calculate_f():
    sum = 0
    sum_squares = 0

    while True:
        num = int(input())
        sum += num
        sum_squares += num ** 2

        if sum == 0:
            break

    return sum_squares

result = calculate_f()
print(result)
```

```python
from sklearn.datasets import fetch_california_housing
import pandas as pd

# Загружаем встроенный датасет
california = fetch_california_housing()

# Создаём DataFrame из признаков
df_california = pd.DataFrame(california.data, columns=california.feature_names)
df_california
```

```python
df_target=pd.DataFrame(california.target, columns=california.target_names)
df_target

df_california_target = pd.concat([df_california, df_target], axis=1)
df_california_target
```
